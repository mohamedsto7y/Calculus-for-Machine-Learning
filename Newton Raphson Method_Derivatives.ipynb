### Use python code to build a function that calculate the approximate solution of a function using Newton-Raphson method.
- Think about function arguments.
- Choose the suitable stop criteria.

### Use your function to solve the following equations:
- $𝒇(𝒙)=𝒙^𝟐−𝒙-𝟏$ , $𝒙_0=1$     
- $𝒇(𝒙)=𝒙^3−7𝒙^2+8x-3$ , $𝒙_0=5$     
- $𝒇(𝒙)=𝒙cos(x)−𝒙^2$ , $𝒙_0=1$     

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import sympy as sp
def newton_raphson_method(func, first_guess, max_iter=20, e=0.000001, decimal_similarity=6):
    i = 0
    xi_1 = first_guess
    derivative = sp.diff(func, x)
    print("iteration:"+str(i)+" ,x ="+  str(xi_1) +" ,f(x) ="+str(func.subs(x, xi_1)))
    while abs(func.subs(x, xi_1)) > e and i < max_iter:
        i += 1
        dfdx = derivative.subs(x, xi_1)
        if dfdx == 0:
            print("error division by zero")
            return None, None
        xi = xi_1 - func.subs(x, xi_1) / dfdx
        print("iteration:"+str(i)+" ,x ="+ "(str(xi_1))" +" ,f(x) ="+str(func.subs(x, xi_1)))
        if abs(xi - xi_1) < 1 / 10**decimal_similarity:
            print(f"stop check,the root is similar to the previous iteration within {decimal_similarity} decimal places.")
            return xi, i
        xi_1 = xi
    if i == max_iter:
        raise ValueError("max itiration reached")
    return xi, i
```

```python
example_function = x**2 - x - 1
first_guess = -1.0
root, iterations = newton_raphson_method(example_function, first_guess)
print("approximate root:", root)
print("number of iterations:", iterations)
```

```python
example_function = x**3 - 7*x **2 + 8*x -3
first_guess = 5.0
root, iterations = newton_raphson_method(example_function, first_guess)
print("approximate root:", root)
print("number of iterations:", iterations)
```

```python
example_function = x * sp.cos(x) - x**2
first_guess = -10.0
root, iterations = newton_raphson_method(example_function, first_guess)
print("approximate root:", root)
print("number of iterations:", iterations)
```

```python
x_vals = np.linspace(-2, 2, 1000)
y_vals = np.array([example_function.subs(x, val) for val in x_vals])
plt.plot(x_vals, y_vals)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
```
